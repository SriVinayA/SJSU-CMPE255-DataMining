## 1. Binary Classification

```python
from pycaret.classification import *

# Load the data
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/SJSU/Sem1/CMPE255 - Data Mining/Assignment 2.2/bmi.csv')

# Convert 'BmiClass' into a binary column
df['BinaryBmiClass'] = df['BmiClass'].apply(lambda x: 1 if x == "Obese Class 1" else 0)

# Setup the environment for binary classification
binary_clf_setup = setup(data=df, target='BinaryBmiClass', session_id=123, use_gpu=True)

# Compare models
binary_clf_compare = compare_models()

print(binary_clf_compare)
```

## 2. Multiclass Classification

```python
from pycaret.classification import *

# Setup the environment for multiclass classification
multi_clf_setup = setup(data=df, target='BmiClass', session_id=123, use_gpu=True)

# Compare models
multi_clf_compare = compare_models()

print(multi_clf_compare)
```
