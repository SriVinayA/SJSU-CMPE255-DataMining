```python
from pycaret.classification import *

# Load the data
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/SJSU/Sem1/CMPE255 - Data Mining/Assignment 2.2/Employee.csv')
df.head()
```

```python
# Setup the environment for binary classification
binary_clf_setup = setup(data=df, target='LeaveOrNot', session_id=123, use_gpu=True)

# Compare models
binary_clf_compare = compare_models()

print(binary_clf_compare)
```

```python
from scipy.sparse import random
import gradio as gr
import random
def predict_genre(Education, JoiningYear, City, PaymentTier, Age, Gender, EverBenched, ExperienceInCurrentDomain):
    # Assuming 'data' is a dictionary with input features

    dic = {'Education':[Education],'JoiningYear': [JoiningYear], 'City': [City], 'PaymentTier':[PaymentTier],'Age': [Age], 'Gender': [Gender], 'EverBenched':[EverBenched],'ExperienceInCurrentDomain': [ExperienceInCurrentDomain]}
    input_data = pd.DataFrame.from_dict(dic)
    result = predict_model(binary_clf_compare, data=input_data)
    prediction = result['prediction_label'][0]  # Extract the predicted label
    return prediction
iface = gr.Interface(
    fn=predict_genre,
    inputs=[
        gr.inputs.Textbox(label="Education"),
        gr.inputs.Textbox(label="JoiningYear"),
        gr.inputs.Textbox(label="City"),
        gr.inputs.Textbox(label="PaymentTier"),
        gr.inputs.Textbox(label="Age"),
        gr.inputs.Textbox(label="Gender"),
        gr.inputs.Textbox(label="EverBenched"),
        gr.inputs.Textbox(label="ExperienceInCurrentDomain")

    ],
     outputs=gr.outputs.Textbox(label="Predicted Leave or Not"),
)
iface.launch(share=True, debug=True)
```
